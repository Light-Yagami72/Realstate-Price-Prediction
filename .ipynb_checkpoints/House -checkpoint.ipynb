<h1>Outlier</h1>
An outlier is a point which falls more than 1.5 times the interquartile range above the third quartile or below the first quartile

Z score is an important concept in statistics. Z score is also called standard score. This score helps to understand if a data value is greater or smaller than mean and how far away it is from the mean. More specifically, Z score tells how many standard deviations away a data point is from the mean.
               <h5>Z score = (x -mean) / std. deviation</h5>



<h3> Detecting Outliner</h3>
<ol>
<li>detecting outliers</li>
<li> z_score</li>
<li>the threshhold=3 will beused to take almost all the data</li>
    </ol>

```python
outlier=[]
def detectoutliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>3:
            outlier.append(i)
    return outlier
```

<h2>Detecting Outliers in Various Columns</h2>

```python
x=detectoutliers(house_df['total_rooms'])
np.min(x)
house_df[house_df['total_rooms']>np.min(x)]
```

```python
x=detectoutliers(house_df['population'])
np.min(x)
house_df[house_df['population']>np.min(x)]
```

```python
x=detectoutliers(house_df['housing_median_age'])
np.min(x)
house_df[house_df['housing_median_age']>np.min(x)]
```

```python
x=detectoutliers(house_df['households'])
np.min(x)
house_df[house_df['households']>np.min(x)]
```

```python
x=detectoutliers(house_df['median_income'])
np.min(x)
house_df[house_df['median_income']>np.min(x)]
```
